```python
#Importazione delle librerie
from urllib import request
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
import json
import pandas as pd
import pymongo
```

### Funzione di web scraping


```python
def get_page(url):
    # add the correct User-Agent
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}

    # open the page
    page_request = request.Request(url, headers=headers)
    return BeautifulSoup(request.urlopen(page_request), 'html.parser')
```

```python
def date_to_string(date):
    return date.strftime("%d/%m/%Y")
```

```python
def write_json_on_file(file_name, dictionary):
    with open(file_name, 'w') as f:
        json.dump(dictionary, f, indent = 4)
```

```python
def print_dict(event):
    for key in event:
        print(f'''{key}: {event[key]}''')
    print()
```

```python
base_url= "https://www.basketball-reference.com/boxscores/?"

date = datetime.today()
```

```python
data = []
while(not(date.day == 29 and date.month == 10 and date.year == 2000)):
    url = base_url + "month=" + str(date.month) + "&day=" + str(date.day) + "&year=" + str(date.year)
    html_content = get_page(base_url + "month=" + str(date.month) + "&day=" + str(date.day) + "&year=" + str(date.year))

    game_summaries = html_content.select(".game_summary")
    for game_summary in game_summaries:
        try:
            game_summary_url = "https://www.basketball-reference.com" + game_summary.select(".links")[0].select("a")[0]["href"]
            game_summary_page = get_page(game_summary_url)

            scorebox = game_summary_page.select(".scorebox .logo + strong a")
            
            team_away = scorebox[0].text
            team_home = scorebox[1].text

            try:
                attendance = game_summary_page.select("#content > div")[-2].select("div")[2].text
                attendance = int(attendance.replace("Attendance:", "").replace(",", ""))
            except:
                attendance = 0

            event = {
                "date": date_to_string(date),
                "teamAway":team_away,
                "teamHome": team_home,
                "attendance": attendance
            }
            
            data.append(event)

            print_dict(event)
        except:
            continue

    date -= timedelta(hours=24)
```

```python
#Salvataggio file in formato csv
write_json_on_file("nba_attendance.json", data)
pd.DataFrame.from_dict(data).to_csv("nba_attendance_vero.csv")
```
